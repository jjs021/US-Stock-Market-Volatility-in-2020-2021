```python
import altair as alt
from vega_datasets import data

import pandas as pd
```

```python
# Since the data is more than 5,000 rows we'll import it from a URL

df = pd.read_csv("SP10yrPriceHistory.csv")
df2 = pd.read_csv("SPPriceHistory2021.csv")
```

```python
df.Date = pd.to_datetime(df.Date)
```

```python
df2.Date = pd.to_datetime(df2.Date)
```

```python
df2.dtypes
```

```python
df2.head()
```

```python
df['month'] = pd.DatetimeIndex(df.Date).month
df['day'] = pd.DatetimeIndex(df.Date).day
df.head(10)
```

```python
df2['month'] = pd.DatetimeIndex(df2.Date).month
df2['day'] = pd.DatetimeIndex(df2.Date).day
df2.head(10)
```

```python
source = df

alt.Chart(
    source,
    title="S&P 500 Performance in 2020"
).mark_rect().encode(
    x='day:O',
    y='month:O',
    color=alt.Color('% Change', scale=alt.Scale(scheme="redblue")),
    tooltip=[
        alt.Tooltip('Date:T', title='Date'),
        alt.Tooltip('% Change:Q', title='Change in perentage')
    ]
).properties(width=750)
```

```python
source = df2

alt.Chart(
    source,
    title="S&P 500 Performance in 2021"
).mark_rect().encode(
    x='day:O',
    y='month:O',
    color=alt.Color('% Change', scale=alt.Scale(scheme="redblue")),
    tooltip=[
        alt.Tooltip('Date:T', title='Date'),
        alt.Tooltip('% Change:Q', title='Change in perentage')
    ]
).properties(width=750)
```

The U.S. stock market had more extreme up and down days in 2021 than in 2020.
